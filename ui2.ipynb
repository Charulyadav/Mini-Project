```python
import gradio as gr
import numpy as np
import tensorflow as tf
from PIL import Image
```

```python
# Load your model
model = tf.keras.models.load_model("C://Users//Charul//Desktop//MCA'24//Mini Project//Image-Forgery-Detection-master//model_casia_run1.h5")
```

```python
model.summary()
```

```python
def classify_image(file):
    img = Image.open(file.name)
    img_array = np.array(img) / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)
    img_array = tf.image.resize(img_array, (128, 128))
    img_array = tf.reshape(img_array, (-1, 128, 128, 3))

    prediction = model.predict(img_array)

    image_component = gr.Image(file.name)

    return [image_component, f"Prediction: {'Real' if prediction[0][0] >= 0.43 else 'Fake'}"]

iface = gr.Interface(fn=classify_image, inputs="file", outputs=["image", "text"])
iface.launch(share=True, inbrowser=True)
```
